# 크롤링

```python
#!/usr/bin/python
# -*- coding: UTF-8 -*-

from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import csv
import os
import random
import time

def parse_div(div, i):
    category = div.select('div.flex-grow-1 > h4 > a.mr-1')[0].text
    title = div.select('div.flex-grow-1 > h4 > span.title.mr-1')[0].text
    rated = div.select('div.flex-grow-1 > h4 > span:nth-child(7)')     
    if rated == []:
        rated = div.select('div.flex-grow-1 > h4 > span:nth-child(6)')
    rated = rated[0].text
    data_id = div.attrs['data-id']
    link = 'https://bgmstore.net/view/' + data_id + '/' + title
    return [category[1:-1], title, rated[1:], link]

def parse_page():
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    div = soup.select('#root > div > div > div > div > main > div > div.container__wrap > div:nth-child(2) > div > div > div.d-flex.with-sidebar-right > div.flex-grow-1.w-100 > div:nth-child(3) > div > div > div > div')

    for i in range(len(div)):
        div_i = div[i]
        filename ="C:/Users/s4241/Desktop/Competition/alldata.csv"#저장할 csv파일 위치
        os.makedirs(os.path.dirname(filename), exist_ok=True)
        with open(filename, 'a', newline='', encoding='utf-8-sig') as csv_file:
            csv_writer = csv.writer(csv_file)
            row = parse_div(div_i, i)
            csv_writer.writerow(row)

def crawling():
    for i in range(239):
        if i > 223:
            ## url에 접근
            driver.get('https://bgmstore.net/search?page=' + str(i) + '&q=%EC%9E%90%EC%9E%91%EA%B3%A1&qType=category')
            #sleep_time = random.random()*0.1
            #time.sleep(1)
            parse_page()
        
## chromedriver의 위치를 지정
driver = webdriver.Chrome('C:/chromedriver')
driver.implicitly_wait(100)

crawling() 

driver.close()
```
